6. Explain the concept of generators in Python and how they are defined.
   Generators are iterators created using functions with yield.

They don’t store all values in memory; they produce them one by one.

```python
def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1
```

```python
for num in count_up_to(3):
    print(num)
```

10. What is the difference between map() and filter() functions in Python?
    map(func, iterable)	-- Applies func to every element	-- Transformed values
filter(func, iterable)	-- Keeps elements where func returns True	-- Filtered subset

```python
nums = [1, 2, 3, 4]
print(list(map(lambda x: x*2, nums)))      # [2, 4, 6, 8]
print(list(filter(lambda x: x%2==0, nums)))  # [2, 4]
```

```python
# 4. Function to check if number is prime (1 to 200)

def isPrime(n):
    if n < 2:
        return False
        
    for i in range(2, int(n ** 0.5)+1):
        if n % i == 0:
            return False

    return True

primes = [n for n in range(1,201) if isPrime(n)]
print(primes)
```

5. Iterator class for Fibonacci sequence

```python
class Fibonacci:
    def __init__(self, n):
        self.n = n
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.n:
            result = self.a
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return result
        else:
            raise StopIteration

for num in Fibonacci(10):
    print(num, end=" ")  
```

6. Generator function for powers of 2

```python
def power_of_two(limit):
    for exp in range(limit+1):
        yield 2 ** exp

for num in power_of_two(5):
    print(num, end=" ")
```

```python
# 7. Generator function to read file line by line
def read_file_line(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.strip()

for line in read_file_line('sample.txt'):
    print(line)
```

10. Using filter() to remove vowels from a string

```python
string = "tejas"
def rem_vowels(string):
    vowels = "aeiouAEIOU"
    return ''.join(filter(lambda ch:ch not in vowels, string))
```

```python
rem_vowels(string)
```
